```python
### Getting fasta sequences in a loop
### Retriving fasta sequences for the ids generated for each category for all 20 genotypes
###
def getting_fasta(genotype,sequence_file):
    main_path = 'C:\\Users\\johnp\\Downloads\\poly-a_loop\\forward_strand\\Genotypes\\'
    for i in range(1,9):

        filename  = main_path + genotype +"\\ids\\Category_%i.txt"%i ## Getting path for the each category file
        file_ids = open(filename,"r") ## Opening the file

        filename_fasta = main_path + genotype + "\\fasta\\Category_%i.txt"%i ## Getting path for each category file that contain fasta sequences
        re = open(filename_fasta,"a")  ## Creating file for category_x.txt to store fasta sequences         
        ids = []
        fasta = []
        for line in file_ids: ### Looping through each header in Category_x.txt file
            line = line.strip("\n") ## Remove "\n" character
            ids.append(line) ## Storing header ids in a list
        # print(ids)
        ## Getting path for initial fasta files for 20 genotypes
        main_path_to_sequence_file = "C:\\Users\\johnp\\Downloads\\poly-a_loop\\forward_strand\\intial_fasta_files\\"
        sequence_path =  main_path_to_sequence_file + sequence_file
        file_fasta = open(sequence_path,"r")
        ids_seq_list = []
        for line2 in file_fasta: ## Looping through each line in a fasta file
            words = line2.split() ## Converting words in each line to a list
            if line2[0] == ">": ## If a line starts with ">"
                    y = line2 ## Store header in y variable
            else: ## If line doesn't starts with ">"
                    line2 = line2.strip("\n") ## There is "\n" character after each header, we are just getting rid of it
                    ids_seq_list.append((y.strip("\n"),line2)) ## List that contains tuple of (>Header and the fasta sequence)
        # print(ids_seq_list)

        ## Getting fasta sequences after comparing ids
        for x in ids: ## Looping through a list that contains ids
        #     print("Working")
            for y in ids_seq_list: ## Looping through a list that contains a tuple of (ids and sequences)
        #         print("Working")
                y1 = y[0] ## Storing id value in (ids and sequences) to a variable y1
                if (x == y1): ## If an id in ids_list matches ids_in (ids and sequences)
    #                 print(x + "\n" + y[1])
                    re.write(x) ## Writing header to a file
                    re.write("\n") ## New line
                    re.write(y[1]) ## Writing fasta sequence assosiated with that id
                    re.write("\n") ## New line

    re.close()
    
    return 0
## At the end you will get 8 different Categories named Category_1.txt, Category_2,.....So on.... Category_8.txt that
# will contain fasta sequences
```
