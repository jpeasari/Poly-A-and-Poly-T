```python
### Getting fasta sequences in a loop
def getting_fasta(genotype,sequence_file):
    main_path = 'C:/Users/johnp/Downloads/poly-a_loop/reverse_strand/Genotypes/'
    for i in range(1,9):

        filename  = main_path + genotype +"/ids/Category_%i.txt"%i
        file_ids = open(filename,"r")

        filename_fasta = main_path + genotype + "/fasta/Category_%i.txt"%i
        re = open(filename_fasta,"a")           
        ids = []
        fasta = []
        for line in file_ids: ### Working with sequence header only
            line = line.strip("\n")
            ids.append(line)
        # print(ids)
        main_path_to_sequence_file = "C:/Users/johnp/Downloads/poly-a_loop/reverse_strand/intial_fasta_files/"
        sequence_path =  main_path_to_sequence_file + sequence_file
        file_fasta = open(sequence_path,"r") ### Tuple with id and seq
        ids_seq_list = []
        for line2 in file_fasta:
            words = line2.split()
            if line2[0] == ">":
                    y = line2
            else:
                    line2 = line2.strip("\n") 
                    ids_seq_list.append((y.strip("\n"),line2))
        # print(ids_seq_list)

        ## Getting fasta sequences after comparing ids
        for x in ids:
        #     print("Working")
            for y in ids_seq_list:
        #         print("Working")
                y1 = y[0]
                if (x == y1):
    #                 print(x + "\n" + y[1])
                    re.write(x)
                    re.write("\n")
                    re.write(y[1])
                    re.write("\n")

    re.close()
    
    return 0
```
